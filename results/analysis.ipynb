```python
#for_all_og files
def all_og(file_in, file_out):
    og_key = open(file_in).read().splitlines()
    prot = [og[k] for k in og_key]
    x = [i.strip().split() for i in prot]

    prot=[]
    for k in range(len(x)):
        prot += [i for i in x[k]]

    s=[]
    for k in range(len(x)):
        s += [i + "_" + og_key[k] for i in x[k]]

    #check if have ARE
    hy_low = open("hydra_low.txt").read().splitlines()
    hy_low = [i +".t1" for i in hy_low]
    dm_low = open("dmel_low.txt").read().splitlines()
    hs_low = open("hsap_low.txt").read().splitlines()


    p = set(prot) & set(hy_low) | set(prot) & set(dm_low) | set(prot) & set(hs_low)

    d = {k: c[k] for k in prot}
    lines = []
    for i,k in enumerate(prot):
        if(k in p):
            lines.append(">" + s[i] + "_ARE" + '\n')
        else:
            lines.append(">" + s[i] + '\n')
        lines.append(d[prot[i]] + '\n')
    open(file_out,"w").writelines(lines)
```
